```python
with open('10_02_happiness.csv', 'r') as f:
    reader = csv.DictReader(f)
    data = list(reader)
```

```python
print(data[0:10])
```

```python
data = [row for row in data if row['What.is.your.annual.household.income'] != 'NA']
len(data)
```

```python
with open('10_02_happiness_output.csv', 'w') as f:
    writer = csv.writer(f, delimiter='|')
    for row in data:
        writer.writerow([row['How.happy.do.you.feel.right.now'], row['How.satisfied.are.you.with.your.life.in.general'], row['What.is.your.annual.household.income']])
        
```
