## CSV

### Reading

```python
with open('10_02_happiness.csv', 'r') as f:
    reader = list(csv.reader(f, delimiter=','))
    for row in reader[1:]:
        print(row)
```

```python
with open('10_02_happiness.csv', 'r') as f:
    reader = list(csv.DictReader(f, delimiter=','))
    for row in reader[:10]:
        print(row)
```

```python
# What.is.your.annual.household.income

with open('10_02_happiness.csv', 'r') as f:
    reader = list(csv.DictReader(f, delimiter=','))
    data = [row for row in reader if row['What.is.your.annual.household.income'] != 'NA']
    
len(data)
```

### Writing

```python
with open('10_02_happiness_output.csv', 'w') as f:
    writer = csv.writer(f)
    for row in data:
        writer.writerow([row['What.is.your.annual.household.income'], row['How.happy.do.you.feel.right.now']])
    
```
