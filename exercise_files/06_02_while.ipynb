```python
from datetime import datetime
```

```python
datetime.now().second
```

```python
wait_until = 55
while datetime.now().second != wait_until:
    pass
print(f'We are at {wait_until} seconds!')
```

```python
wait_until = 55
while True:
    if datetime.now().second == wait_until:
        print(f'We are at {wait_until} seconds!')
        break
```

```python
wait_until = datetime.now().second + 2

while datetime.now().second != wait_until:
    print(f'We are at {wait_until} seconds!')
    break
```

```python
wait_until = datetime.now().second + 2
while datetime.now().second != wait_until:
    print('Foo!')
    continue
    print(f'We are at {wait_until} seconds!')
    break
    
    
```
